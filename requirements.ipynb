```python
#Requirements
#libraries
numpy
pandas
matplotlib
yfinance
scikit-learn
tensorflow

Project Abstract
Title: Stock Market Price Prediction Using Long Short-Term Memory (LSTM)

Objective: The primary goal of this project is to develop a machine learning model 
capable of predicting the future closing prices of a specific stock based on historical data.
Given the volatile and non-linear nature of financial markets, this project utilizes Deep Learning techniques
to identify patterns that traditional linear models might miss.

Methodology: The project implements a Recurrent Neural Network (RNN) specifically using the Long Short-Term Memory (LSTM) architecture.
LSTMs are chosen for their ability to maintain "memory" over long sequences, making them ideal for time-series forecasting.
We use the yfinance API to fetch real-time historical data, normalize the features using MinMaxScaler, 
and train the model on a look-back window of 60 days to predict the price of the 61st day.
```
